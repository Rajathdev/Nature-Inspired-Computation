```python
import random
import numpy as np


class Perfect_Solution:
    
    def __init__(self,no_of_bins,list_of_item_weight,chromosome_given=None):
        
        
        
        self.no_of_bins, self.list_of_item_weight=no_of_bins, list_of_item_weight
        
        if chromosome_given is None:
            #write method to generate chromosome
            self.chromosome_given =self.generate_random_chromosome()
        else:
            self.chromosome_given=chromosome_given
        #write a method to calculate fitness
        self.count=0
        self.fitness_calculation_of_chromosome()
    
    def _repr_(self):
        return f"Perfect_Solution(fitness={self.fitness_of_chromosome})"
            
    def generate_random_chromosome(self):
        
        return np.random.randint(low=1,high=self.no_of_bins+1,size=len(self.list_of_item_weight))

    def mutate(self,number_of_mutations):

        for i in range(number_of_mutations):
           index_to_mutate=random.randint(0,len(self.chromosome_given)-1)
           self.chromosome_given[index_to_mutate]=random.randint(1,self.no_of_bins+1)


    @property
    def iter_count(self):
        """Getter for counting how many evaluations have taken place for this instance.

        Property decorator stops the user from incorrectly changing the iter count,
        as this has implications for the running of the outer algorithm.

        Returns:
            int: The number of times the solution has been evaluated
        """
        return self.count
       

    def fitness_calculation_of_chromosome(self):
        
        self.count+=1
        #bins_and_items=counter();
        
        bins_and_items={}
        for index,bin_position_of_item in enumerate(self.chromosome_given):
            if bin_position_of_item not in bins_and_items:
                bins_and_items[bin_position_of_item]=0
                
            bins_and_items[bin_position_of_item]+=self.list_of_item_weight[index] 
            
        self.fitness_of_chromosome=max(bins_and_items.values())-min(bins_and_items.values())
    
    
    
def select_a_parent(pop):
    parent_one=binary_tournment_to_select_parent(pop)

    parent_two=binary_tournment_to_select_parent(pop)

    return parent_one,parent_two


def select_a_parent_min_fitness(pop):
    
    chromosome_fitness=[]
    for i in pop:
        chromosome_fitness.append(i.fitness_of_chromosome)

    value_min=min(chromosome_fitness)

    chromosome_fitness.remove(value_min)
    value_second_min=min(chromosome_fitness)
    for i in pop:
        if i.fitness_of_chromosome==value_min:
            parent_one=i
        if i.fitness_of_chromosome==value_second_min:
            parent_two=i
    
    return parent_one,parent_two


def binary_tournment_to_select_parent(pop):

    random_chromosome=random.choices(pop, k=2)
    chromosome_fitness=[]
    for i in random_chromosome:
        chromosome_fitness.append(i.fitness_of_chromosome)

        value_min=min(chromosome_fitness)   
    for i in random_chromosome:
        if i.fitness_of_chromosome==value_min:
            return i
         


def count_iterations(all_sol):
    """From a list of solutions, count the number of times they have all been 
       evaluated.

    Args:
        all_solutions: List consisiting of every solution that has ever been
                       in the population.
        
    Returns:
        int: The sum of all iteration counts for each member of the list

    """

    return sum(solution.count for solution in all_sol)

def crossover_of_parent_chromosome(parent_one,parent_two):

    random_crossover_index=random.randint(1,len(parent_one.chromosome_given)-2)
    #print(parent_one.chromosome_given)

    a1 = parent_one.chromosome_given[:random_crossover_index]
    a2 = parent_one.chromosome_given[random_crossover_index:]
    
    b1 = parent_two.chromosome_given[:random_crossover_index]
    b2 = parent_two.chromosome_given[random_crossover_index:]

    # Create two new children from the crossover
    child_one = Perfect_Solution(no_of_bins=parent_one.no_of_bins, list_of_item_weight=parent_one.list_of_item_weight, chromosome_given=np.hstack((a1, b2)))
    child_two = Perfect_Solution(no_of_bins=parent_one.no_of_bins, list_of_item_weight=parent_one.list_of_item_weight, chromosome_given=np.hstack((b1, a2)))

    #child_one=parent_one.chromosome_given[:random_crossover_index]+parent_two.chromosome_given[random_crossover_index:]
    #child_two=parent_two.chromosome_given[:random_crossover_index]+parent_one.chromosome_given[random_crossover_index:]

    
    return child_one,child_two

def crossover_of_parent_chromosome1(parent_one,parent_two):
    random_crossover_index=random.randint(1,len(parent_one.chromosome_given)-2)
    allowed_values = list(range(1,len(parent_two.chromosome_given)-1))
    #print(random_crossover_index)
    try:
        allowed_values.remove(random_crossover_index)
    except:
        print(random_crossover_index)
        
    #r=range(1,random_crossover_index) + range(random_crossover_index+1,len(parent_one.chromosome_given)-2)
    random_crossover_index2= random.choice(allowed_values)
    #print(random_crossover_index,random_crossover_index2)
    while random_crossover_index==random_crossover_index2:
        random_crossover_index2=random.randint(1,len(parent_one.chromosome_given)-2)
        print("equal")   
    if random_crossover_index<random_crossover_index2 :
        #print("greater")
        a1 = parent_one.chromosome_given[:random_crossover_index]
        a2 = parent_one.chromosome_given[random_crossover_index:random_crossover_index2]
        a3 = parent_one.chromosome_given[random_crossover_index2:]

        b1 = parent_two.chromosome_given[:random_crossover_index]
        b2 = parent_two.chromosome_given[random_crossover_index:random_crossover_index2]
        b3 = parent_two.chromosome_given[random_crossover_index2:]
    else:
        #print("lesseer")
        a1 = parent_one.chromosome_given[:random_crossover_index2]
        a2 = parent_one.chromosome_given[random_crossover_index2:random_crossover_index]
        a3 = parent_one.chromosome_given[random_crossover_index:]

        b1 = parent_two.chromosome_given[:random_crossover_index2]
        b2 = parent_two.chromosome_given[random_crossover_index2:random_crossover_index]
        b3 = parent_two.chromosome_given[random_crossover_index:]
    # Create two new children from the crossover
    child_one = Perfect_Solution(no_of_bins=parent_one.no_of_bins, list_of_item_weight=parent_one.list_of_item_weight, chromosome_given=np.hstack((a1, b2,a3)))
    child_two = Perfect_Solution(no_of_bins=parent_one.no_of_bins, list_of_item_weight=parent_one.list_of_item_weight, chromosome_given=np.hstack((b1, a2,b3)))
    #print(len(child_one.chromosome_given),len(child_two.chromosome_given))
    return child_one,child_two    

def replace_weakest(pop,child_after_mutation):
    
    pop_fitness=[]

    for i in pop:
        pop_fitness.append(i.fitness_of_chromosome)

    worst_fitness=max(pop_fitness)
    #print("rajath",worst_fitness)
    if child_after_mutation.fitness_of_chromosome <= worst_fitness:
        for i in pop:
            if i.fitness_of_chromosome== worst_fitness:
                pop.remove(i)
                pop.append(child_after_mutation)

    return pop



def evol_sol(no_of_bins,list_of_item_weight,no_of_mutation,population_size,crossover_of_chromosome=True,no_of_iterations=10000):

    pop=[]
    for i in range(population_size):
        pop.append(Perfect_Solution(no_of_bins,list_of_item_weight))
    
    intial_fitness=[]
    for i in pop:
        intial_fitness.append(i.fitness_of_chromosome)

    avg_pop= sum(intial_fitness)/population_size 

    all_sol=pop[:]



    while True:

        if count_iterations(all_sol) >= no_of_iterations:
            break

        if crossover_of_chromosome:
            parent_one,parent_two= select_a_parent(pop)
            #parent_one,parent_two= select_a_parent_min_fitness(pop)
            child_one,child_two=crossover_of_parent_chromosome(parent_one,parent_two)

            if no_of_mutation>0:
                child_one.mutate(no_of_mutation)
                child_two.mutate(no_of_mutation)

            all_sol+=[child_one,child_two]
            pop=replace_weakest(pop,child_one)
            pop=replace_weakest(pop,child_two)
        else:
            random_candidate=binary_tournment_to_select_parent(pop)
            #random_candidate=random.choices(select_a_parent_min_fitness(pop),k=1)
            #for i in random_candidate:
            child=Perfect_Solution(no_of_bins,list_of_item_weight,chromosome_given=random_candidate.chromosome_given)

            if no_of_mutation>0:
                child.mutate(no_of_mutation)
                

            all_sol+=[child]
            pop=replace_weakest(pop,child)     

    best_solution = min(solution.fitness_of_chromosome for solution in pop)

    return best_solution, avg_pop        

 




def bin_exp(bin1,list_one,bin2,list_two,no_of_mutation,population_size,crossover_of_chromosome=True,no_of_iterations=10000,exper=5):




    bbp1=[]
    bbp2=[]
    for i in range(exper):
        bbp1.append(evol_sol(bin1,list_one,no_of_mutation,population_size,crossover_of_chromosome,no_of_iterations))
        bbp2.append(evol_sol(bin2,list_two,no_of_mutation,population_size,crossover_of_chromosome,no_of_iterations))
        print(bbp1,bbp2)
 
    return bbp1,bbp2



     

                
#if __name__==__main__.Perfect_Solution:
def main():
    list_one=[]
    list_two=[]
    for i in range(1,501):
        list_one.append(float(i/2))
        list_two.append(float((i**2)/2))

    total_experiments=[]

    print("one")
    total_experiments.append(bin_exp(10, list_one, 100, list_two, no_of_mutation=1, population_size=10))
    print("two")
    total_experiments.append(bin_exp(10, list_one, 100, list_two, no_of_mutation=1, population_size=100))
    print("three")
    total_experiments.append(bin_exp(10, list_one, 100, list_two, no_of_mutation=5, population_size=10))
    print("four")
    
    total_experiments.append(bin_exp(10, list_one, 100, list_two, no_of_mutation=5, population_size=100))
    print("five")    
    total_experiments.append(bin_exp(10, list_one, 100, list_two, no_of_mutation=5, population_size=10, crossover_of_chromosome=False))
    print("six")
    total_experiments.append(bin_exp(10, list_one, 100, list_two, no_of_mutation=0, population_size=10))
    

main()
                
        
```
